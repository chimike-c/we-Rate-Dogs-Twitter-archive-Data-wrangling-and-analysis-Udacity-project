In the visualisation section, some graphical representations of data are presented, buttressing the the various insights that were inferred.

The output of the following code blocks give the graphical representations for tweet interactions (retweets, favourites) for various predicted breeds. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
twitter_archive_master = pd.read_csv('twitter_archive_master.csv')
```

```python
interactions_per_breed = twitter_archive_master.groupby("predicted_breed")['retweet_count', 'favorite_count']
```

```python
retweet_interactions_per_breed = interactions_per_breed.sum()['retweet_count'].sort_values(ascending=False).reset_index(name="retweet_count")
```

```python
favorite_interactions_per_breed = interactions_per_breed.sum()['favorite_count'].sort_values(ascending=False).reset_index(name="favorite_count")
```

```python
sns.barplot(y="predicted_breed", x="favorite_count", data=favorite_interactions_per_breed.head(10))
plt.title("Total Favourite Counts");
```

```python
sns.barplot(y="predicted_breed", x="retweet_count", data=retweet_interactions_per_breed.head(10))
plt.title("Total Retweet Counts");
```

The barplots show the dog breeds that had the highest user interactions (retweets and favorites). The results show that users interacted most with tweets about golden retrievers, Labrador retrievers, Pembroke and Chihuahuas.
